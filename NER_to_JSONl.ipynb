```python
from tabulate import tabulate
import pandas as pd
from itertools import zip_longest
import json
import numpy as np


#grouped_df=grouped_df.head(2)
# Assuming df is your DataFrame
data = grouped_df.copy()

df = pd.DataFrame(data)

# Create a mapping of word, entity_group, startOffset, and endOffset
word_mapping = []

# Iterate through the DataFrame
for _, row in df.iterrows():
    words = row['word'].split("^")
    entity_groups = row['entity_group']

    for word, entity_group in zip_longest(words, entity_groups, fillvalue='others'):
        start_offset = sum(len(w) + 1 for w in words[:words.index(word)]) if words.index(word) > 0 else 0
        end_offset = start_offset + len(word)

        word_mapping.append({
            'word': word,
            'entity_group': entity_group,
            'startOffset': start_offset,
            'endOffset': end_offset
        })
    # Create a new row as a dictionary
    new_row = {'word': "xxx",'entity_group': "xxx",'startOffset': "xxx",'endOffset': "xxx"}

    # Append the new row to the DataFrame
    word_mapping.append(new_row)

# Update startOffset and endOffset to 0 where entity_group is "others"
for mapping in word_mapping:
    if mapping['entity_group'] == 'others':
        mapping['startOffset'] = 0
        mapping['endOffset'] = 0

# Create a DataFrame from the mapping
word_mapping_df = pd.DataFrame(word_mapping)
word_mapping_df.to_csv('word_mapping_df.csv')

# Print only the required columns using tabulate
columns_to_print = ['word', 'entity_group', 'startOffset', 'endOffset']
print(tabulate(word_mapping_df[columns_to_print], headers='keys', tablefmt='pretty'))
```

```python
prev = 0
with open('output55.jsonl', 'w') as jsonl_file:
  for i in range(word_mapping_df.shape[0]):
    final_d = {}
    filtered_annotations = []
    if word_mapping_df['word'].iloc[i] == "xxx" :
      current = i
      temp = word_mapping_df[prev:current]
      df = temp[temp['entity_group'] != 'others']
      l = df.values.tolist()

      for entity in l:
        d =  {}
        d['endOffset'] = entity[3]
        d['startOffset'] = entity[2]
        d['displayName'] = entity[1]
        filtered_annotations.append(d)
      final_d["text_segment_annotations"]= filtered_annotations
      final_d["textContent"] = ' '.join(temp["word"])

      #print(final_d)
      jsonl_file.write(json.dumps(final_d) + '\n')
      prev = i+1
```

```python
output_file_path="output55.jsonl"
# Print the content of the JSONL file
with open(output_file_path, 'r') as jsonl_file:
    print(f"Content of '{output_file_path}':\n")
    for line in jsonl_file:
        print(line.strip())
```
